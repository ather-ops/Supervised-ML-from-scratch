# Naive Bayes - Day 1 (From Scratch)

## What is Naive Bayes?

Naive Bayes is a machine learning classification algorithm that predicts the category of a data point using probability. It assumes that all features are independent of each other. Naive Bayes performs well in many real-world applications such as spam filtering, document categorisation and sentiment analysis.

---

## Why it is Called Naive Bayes?

It is named as **"Naive"** because it assumes the presence of one feature does not affect other features. The **"Bayes"** part of the name refers to its basis in Bayes' Theorem.

---

## Key Features of Naive Bayes Classifiers

| Feature | Description |
|---------|-------------|
| Probabilistic | Uses Bayes' Theorem to classify data |
| Fast | Very few parameters, fast prediction |
| Independent | Assumes each feature is independent of others |
| Versatile | Used in spam filtration, sentiment analysis, article classification |

---

## Assumptions of Naive Bayes

| Assumption | Description |
|------------|-------------|
| Feature Independence | Each feature does not affect any other feature |
| Normal Distribution | Continuous features are normally distributed |
| Multinomial Distribution | Discrete features have multinomial distributions |
| Equal Importance | All features contribute equally |
| No Missing Data | Data should not contain missing values |

---

## Bayes' Theorem

$$P(y|X) = \frac{P(X|y) \cdot P(y)}{P(X)}$$

Where:
- $P(y|X)$: Posterior probability (probability of class y given features X)
- $P(X|y)$: Likelihood (probability of features X given class y)
- $P(y)$: Prior probability of class y
- $P(X)$: Marginal likelihood or evidence

---

## Naive Assumption

$$P(x_1, x_2, ..., x_n | y) = P(x_1 | y) \cdot P(x_2 | y) \cdots P(x_n | y)$$

Thus, Bayes' theorem becomes:

$$P(y | x_1, ..., x_n) \propto P(y) \cdot \prod_{i=1}^{n} P(x_i | y)$$

---

## Naive Bayes Classifier Formula

$$\hat{y} = \arg\max_{y} P(y) \cdot \prod_{i=1}^{n} P(x_i | y)$$

---

## Types of Naive Bayes

| Type | Use Case |
|------|----------|
| Gaussian Naive Bayes | Continuous features with normal distribution |
| Multinomial Naive Bayes | Text classification with term frequencies |
| Bernoulli Naive Bayes | Binary features (word presence/absence) |

---

## Advantages

- Easy to implement and computationally efficient
- Effective with large number of features
- Performs well with limited training data
- Works well with categorical features

---

## Disadvantages

- Assumes features are independent (may not hold in real-world)
- Can be influenced by irrelevant attributes
- May assign zero probability to unseen events

---

## One-Line Summary

**Naive Bayes uses Bayes' Theorem with independence assumption to compute probabilities for classification.**

## Dataset: Golf Playing Decision

```python
# Create the golf dataset
data = {
    'Outlook': ['Rainy', 'Rainy', 'Overcast', 'Sunny', 'Sunny', 'Sunny', 'Overcast', 
                'Rainy', 'Rainy', 'Sunny', 'Rainy', 'Overcast', 'Overcast', 'Sunny'],
    'Temperature': ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 
                    'Mild', 'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild'],
    'Humidity': ['High', 'High', 'High', 'High', 'Normal', 'Normal', 'Normal', 
                 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'High'],
    'Windy': ['False', 'True', 'False', 'False', 'False', 'True', 'True', 
              'False', 'False', 'False', 'True', 'True', 'False', 'True'],
    'Play Golf': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 
                  'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']
}

df = pd.DataFrame(data)
print(df)
```

```python
# Count class distribution
print("\nClass Distribution:")
print(df['Play Golf'].value_counts())
```

## Naive Bayes Implementation from Scratch

```python
class NaiveBayesFromScratch:
    """
    Naive Bayes Classifier from Scratch
    """
    
    def __init__(self):
        self.classes = None
        self.class_priors = {}
        self.feature_probs = {}
    
    def fit(self, X, y):
        """
        Train the Naive Bayes classifier
        """
        self.classes = np.unique(y)
        n_samples = len(y)
        
        # Calculate class priors P(y)
        for cls in self.classes:
            self.class_priors[cls] = np.sum(y == cls) / n_samples
        
        # Calculate conditional probabilities P(x_i | y)
        for cls in self.classes:
            self.feature_probs[cls] = {}
            cls_data = X[y == cls]
            
            for feature_idx in range(X.shape[1]):
                feature_values = cls_data[:, feature_idx]
                unique_vals, counts = np.unique(feature_values, return_counts=True)
                
                # Laplace smoothing to avoid zero probabilities
                probs = {}
                total_count = len(feature_values)
                for val in unique_vals:
                    probs[val] = (np.sum(feature_values == val) + 1) / (total_count + len(unique_vals))
                
                self.feature_probs[cls][feature_idx] = probs
    
    def predict_single(self, x):
        """
        Predict a single sample
        """
        posteriors = {}
        
        for cls in self.classes:
            # Start with prior
            posterior = self.class_priors[cls]
            
            # Multiply by conditional probabilities
            for feature_idx, feature_val in enumerate(x):
                if feature_val in self.feature_probs[cls][feature_idx]:
                    posterior *= self.feature_probs[cls][feature_idx][feature_val]
                else:
                    # Laplace smoothing for unseen values
                    posterior *= 1 / (len(self.feature_probs[cls][feature_idx]) + 1)
            
            posteriors[cls] = posterior
        
        # Return class with highest posterior
        return max(posteriors, key=posteriors.get)
    
    def predict(self, X):
        """
        Predict multiple samples
        """
        return np.array([self.predict_single(x) for x in X])
```

## Training the Model

```python
# Prepare features and target
X = df[['Outlook', 'Temperature', 'Humidity', 'Windy']].values
y = df['Play Golf'].values

# Train Naive Bayes
nb = NaiveBayesFromScratch()
nb.fit(X, y)

print("\nClass Priors:")
for cls, prior in nb.class_priors.items():
    print(f"P({cls}) = {prior:.4f}")
```

## Making Predictions

```python
# Predict on training data
predictions = nb.predict(X)
accuracy = np.mean(predictions == y)

print("\nPredictions:")
for i in range(len(X)):
    print(f"Actual: {y[i]:<5} Predicted: {predictions[i]:<5} {'✓' if predictions[i] == y[i] else '✗'}")

print(f"\nTraining Accuracy: {accuracy*100:.2f}%")
```

## Prediction for New Sample

```python
# New sample: Sunny, Hot, Normal, False
new_sample = np.array(['Sunny', 'Hot', 'Normal', 'False'])
prediction = nb.predict_single(new_sample)

print("\nNew Sample Prediction:")
print(f"Features: Outlook=Sunny, Temperature=Hot, Humidity=Normal, Windy=False")
print(f"Predicted Class: {prediction}")
```
