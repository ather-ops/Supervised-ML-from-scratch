# Gaussian Naive Bayes - Day 2 (From Scratch)

## What is Gaussian Naive Bayes?

Gaussian Naive Bayes is a type of Naive Bayes method working on continuous attributes and the data features that follows Gaussian distribution throughout the dataset. This "naive" assumption simplifies calculations and makes the model fast and efficient.

---

## Key Characteristics

- Works well with continuous numerical features
- Fast, simple, and effective for classification problems
- Commonly used in spam detection, medical diagnosis, and pattern recognition
- Performs well even with small datasets
- Easy to implement and interpret

---

## Mathematics of Gaussian Naive Bayes

Gaussian Naive Bayes assumes that the likelihood $P(x_i | y)$ follows the Gaussian Distribution for each $x_i$ within $y_k$.

$$P(x_i | y) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}$$

Where:
- $x_i$ is the feature value
- $\mu$ is the mean of the feature values for a given class $y_k$
- $\sigma$ is the standard deviation of the feature values for that class
- $\pi$ is a constant (approximately 3.14159)
- $e$ is the base of the natural logarithm

---

## Classification Rule

To classify each new data point x, the algorithm finds the maximum value of the posterior probability of each class and assigns the data point to that class.

$$\hat{y} = \arg\max_{y} P(y) \cdot \prod_{i=1}^{n} P(x_i | y)$$

---

## Why Gaussian Naive Bayes Works Well for Continuous Data?

Gaussian Naive Bayes is effective for continuous data because it assumes each feature follows a Gaussian (normal) distribution. When this assumption holds true, the algorithm performs well in tasks like:
- Medical diagnosis
- Predicting house prices
- Features like age, income, or measurements that follow normal distribution

---

## One-Line Summary

**Gaussian Naive Bayes assumes continuous features follow normal distribution and uses Gaussian PDF to compute likelihoods.**

## Gaussian Naive Bayes Implementation from Scratch

```python
class GaussianNaiveBayes:
    """
    Gaussian Naive Bayes Classifier from Scratch
    """
    
    def __init__(self):
        self.classes = None
        self.class_priors = {}
        self.means = {}
        self.variances = {}
    
    def fit(self, X, y):
        """
        Train the Gaussian Naive Bayes classifier
        """
        self.classes = np.unique(y)
        n_samples = len(y)
        
        # Calculate class priors P(y)
        for cls in self.classes:
            self.class_priors[cls] = np.sum(y == cls) / n_samples
        
        # Calculate mean and variance for each feature per class
        for cls in self.classes:
            X_cls = X[y == cls]
            self.means[cls] = np.mean(X_cls, axis=0)
            self.variances[cls] = np.var(X_cls, axis=0) + 1e-9  # Add small value to avoid division by zero
    
    def _gaussian_pdf(self, x, mean, var):
        """
        Calculate Gaussian Probability Density Function
        """
        return (1 / np.sqrt(2 * np.pi * var)) * np.exp(-((x - mean) ** 2) / (2 * var))
    
    def predict_single(self, x):
        """
        Predict a single sample
        """
        posteriors = {}
        
        for cls in self.classes:
            # Start with prior
            posterior = self.class_priors[cls]
            
            # Multiply by Gaussian likelihood for each feature
            for i in range(len(x)):
                likelihood = self._gaussian_pdf(x[i], self.means[cls][i], self.variances[cls][i])
                posterior *= likelihood
            
            posteriors[cls] = posterior
        
        # Return class with highest posterior
        return max(posteriors, key=posteriors.get)
    
    def predict(self, X):
        """
        Predict multiple samples
        """
        return np.array([self.predict_single(x) for x in X])
    
    def score(self, X, y):
        """
        Calculate accuracy
        """
        return np.mean(self.predict(X) == y)
```

## Example: Petal Length Classification

```python
# Example data: Petal Length and Class
X_example = np.array([1.4, 1.3, 1.5, 4.5, 4.7, 4.6]).reshape(-1, 1)
y_example = np.array([0, 0, 0, 1, 1, 1])

print("Example Dataset:")
for i in range(len(X_example)):
    class_name = "Iris-setosa" if y_example[i] == 0 else "Iris-versicolor"
    print(f"Petal Length: {X_example[i][0]:.1f} -> {class_name}")
```

```python
# Train Gaussian Naive Bayes
gnb_example = GaussianNaiveBayes()
gnb_example.fit(X_example, y_example)

print("\nModel Parameters:")
for cls in gnb_example.classes:
    print(f"\nClass {cls}:")
    print(f"  Mean: {gnb_example.means[cls][0]:.4f}")
    print(f"  Variance: {gnb_example.variances[cls][0]:.4f}")
    print(f"  Prior: {gnb_example.class_priors[cls]:.4f}")
```

```python
# Predict new sample
new_petal_length = np.array([[1.6]])
prediction = gnb_example.predict(new_petal_length)[0]

print("\nNew Sample Prediction:")
print(f"Petal Length: 1.6 cm")
print(f"Predicted Class: {prediction} (Iris-setosa)")
```

```python
# Visualize Gaussian distributions
x_vals = np.linspace(0, 6, 200)

plt.figure(figsize=(10, 6))

for cls in gnb_example.classes:
    mean = gnb_example.means[cls][0]
    var = gnb_example.variances[cls][0]
    std = np.sqrt(var)
    y_vals = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-((x_vals - mean) ** 2) / (2 * var))
    class_name = "Iris-setosa" if cls == 0 else "Iris-versicolor"
    plt.plot(x_vals, y_vals, label=f"Class {cls}: {class_name}")

plt.axvline(x=1.6, color='red', linestyle='--', label='New Sample (1.6)')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Probability Density')
plt.title('Gaussian Distributions for Petal Length')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
