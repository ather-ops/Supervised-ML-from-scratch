# Support Vector Machine (SVM) - Day 1 (From Scratch)

## What is SVM?

Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression tasks. It tries to find the best boundary known as hyperplane that separates different classes in the data.

**Common use cases:** Binary classification like spam vs. not spam or cat vs. dog.

**Main goal:** Maximize the margin between the two classes. The larger the margin, the better the model performs on new and unseen data.

---

## Key Concepts

| Concept | Definition |
|---------|------------|
| Hyperplane | Decision boundary separating different classes in feature space. Equation: wx + b = 0 |
| Support Vectors | Closest data points to the hyperplane, crucial for determining hyperplane and margin |
| Margin | Distance between hyperplane and support vectors. SVM aims to maximize this margin |
| Kernel | Function that maps data to higher-dimensional space for non-linear separation |
| Hard Margin | Perfectly separates data without misclassifications |
| Soft Margin | Allows some misclassifications using slack variables |
| C | Regularization term balancing margin maximization and misclassification penalties |
| Hinge Loss | Loss function penalizing misclassified points or margin violations |

---

## Working of SVM Algorithm

The key idea is to find the hyperplane that best separates two classes by maximizing the margin between them.

**Hard Margin:** The hyperplane that maximizes distance between hyperplane and nearest data points from both classes.

**Soft Margin:** Allows some misclassifications to improve generalization. The objective function balances margin maximization and penalty minimization:

$$\text{Objective} = \frac{1}{\text{margin}} + \lambda \sum \text{penalty}$$

---

## Hinge Loss

- Correctly classified and outside margin: penalty = 0
- Correctly classified but within margin: penalty > 0
- Misclassified: penalty increases proportionally to violation distance

---

## Kernel Trick

When data is not linearly separable, SVM uses kernels to map data into higher-dimensional space.

| Kernel | Description |
|--------|-------------|
| Linear | For linearly separable data |
| Polynomial | Maps data into polynomial space |
| RBF (Radial Basis Function) | Transforms data based on distances between points |

---

## One-Line Summary

**SVM finds the optimal hyperplane that maximizes margin between classes, using kernels for non-linear data.**

## Simple SVM from Scratch (Basic Implementation)

```python
class SimpleSVM:
    """
    Simple SVM implementation from scratch
    """
    
    def __init__(self, learning_rate=0.001, lambda_param=0.01, n_iters=1000):
        self.lr = learning_rate
        self.lambda_param = lambda_param
        self.n_iters = n_iters
        self.w = None
        self.b = None
    
    def fit(self, X, y):
        n_samples, n_features = X.shape
        
        # Convert labels to -1, 1
        y_ = np.where(y <= 0, -1, 1)
        
        # Initialize weights and bias
        self.w = np.zeros(n_features)
        self.b = 0
        
        # Gradient descent
        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                condition = y_[idx] * (np.dot(x_i, self.w) - self.b) >= 1
                if condition:
                    self.w -= self.lr * (2 * self.lambda_param * self.w)
                else:
                    self.w -= self.lr * (2 * self.lambda_param * self.w - np.dot(x_i, y_[idx]))
                    self.b -= self.lr * y_[idx]
    
    def predict(self, X):
        approx = np.dot(X, self.w) - self.b
        return np.sign(approx)
```

```python
# Create simple dataset
X = np.array([[2, 3], [3, 4], [1, 1], [2, 1], [4, 5], [5, 6]])
y = np.array([1, 1, -1, -1, 1, 1])

print("Dataset:")
for i in range(len(X)):
    label = "Class 1" if y[i] == 1 else "Class -1"
    print(f"Point {X[i]} -> {label}")
```

```python
# Train SVM
svm = SimpleSVM(learning_rate=0.001, lambda_param=0.01, n_iters=1000)
svm.fit(X, y)

print("\nTrained SVM:")
print(f"Weights (w): {svm.w}")
print(f"Bias (b): {svm.b}")
```

```python
# Predictions
predictions = svm.predict(X)
print("\nPredictions:")
for i in range(len(X)):
    print(f"Point {X[i]} -> Predicted: {predictions[i]}, Actual: {y[i]}")
```

```python
# Visualize decision boundary
def plot_svm(X, y, svm):
    plt.figure(figsize=(8, 6))
    
    # Plot points
    colors = ['red' if label == 1 else 'blue' for label in y]
    plt.scatter(X[:, 0], X[:, 1], c=colors, s=100, edgecolors='black')
    
    # Plot decision boundary
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))
    
    Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.contourf(xx, yy, Z, alpha=0.2, colors=['blue', 'red'])
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title("SVM Decision Boundary (From Scratch)")
    plt.show()

plot_svm(X, y, svm)
```
