## Understanding Decision Boundaries in KNN

A decision boundary is a line or surface that divides different groups in a classification task. It shows which areas belong to which class based on what the model decides.

---

## Decision Boundaries in KNN

The shape of the boundary depends on:
- The value of K (how many neighbours are considered)
- How the data points are spread out in space

---

## Voronoi Diagrams

For a 1-nearest neighbour (1-NN) classifier, the decision boundary can be visualized using a Voronoi diagram.

A Voronoi diagram splits space into regions based on which training point is closest. Each region called a Voronoi cell contains all the points closest to one specific training point.

---

## How KNN Defines Decision Boundaries

### Impact of 'K' on Decision Boundaries

| k Value | Boundary Shape | Effect |
|---------|----------------|--------|
| Small k | Complex, irregular | Overfitting |
| Large k | Smooth, generalized | Underfitting |

### Impact of Distance Metric

| Metric | Boundary Shape |
|--------|----------------|
| Euclidean | Circular/elliptical boundaries |
| Manhattan | Axis-aligned boundaries |

---

## Factors Affecting KNN Decision Boundaries

| Factor | Effect |
|--------|--------|
| Feature Scaling | Features with larger ranges dominate distance calculations |
| Noise in Data | Outliers can shift or distort boundaries |
| Data Distribution | Influences how KNN separates classes |

---

## One-Line Summary

**KNN decision boundaries are determined by K value and distance metric, with small K creating complex boundaries and large K creating smoother ones.**

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

print("="*50)
print("KNN DAY 6 - DECISION BOUNDARIES")
print("="*50)
```

```python
# KNN from scratch for decision boundary visualization
class KNNScratch:
    def __init__(self, k=3):
        self.k = k
        self.X_train = None
        self.y_train = None
    
    def fit(self, X, y):
        self.X_train = np.array(X)
        self.y_train = np.array(y)
    
    def predict(self, X):
        X = np.array(X)
        predictions = []
        for x in X:
            distances = [np.linalg.norm(x - xt) for xt in self.X_train]
            k_indices = np.argsort(distances)[:self.k]
            k_labels = self.y_train[k_indices]
            predictions.append(Counter(k_labels).most_common(1)[0][0])
        return np.array(predictions)
```

```python
# Create dataset
np.random.seed(42)
X = np.random.randn(200, 2)
y = (X[:, 0] + X[:, 1] > 0).astype(int)

print(f"Dataset shape: {X.shape}")
print(f"Class 0: {np.sum(y == 0)}, Class 1: {np.sum(y == 1)}")
```

```python
# Plot decision boundaries for different k values
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
k_values = [1, 3, 5, 10]

for ax, k in zip(axs.flat, k_values):
    knn = KNNScratch(k=k)
    knn.fit(X, y)
    
    Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    ax.contourf(xx, yy, Z, alpha=0.3, colors=['blue', 'red'])
    ax.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap='bwr', marker='o', s=30)
    ax.set_title(f'KNN Decision Boundaries (k={k})')
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Effect of different distance metrics on decision boundary
class KNNWithMetric:
    def __init__(self, k=3, metric='euclidean'):
        self.k = k
        self.metric = metric
        self.X_train = None
        self.y_train = None
    
    def fit(self, X, y):
        self.X_train = np.array(X)
        self.y_train = np.array(y)
    
    def predict(self, X):
        X = np.array(X)
        predictions = []
        for x in X:
            if self.metric == 'euclidean':
                distances = [np.linalg.norm(x - xt) for xt in self.X_train]
            else:
                distances = [np.sum(np.abs(x - xt)) for xt in self.X_train]
            
            k_indices = np.argsort(distances)[:self.k]
            k_labels = self.y_train[k_indices]
            predictions.append(Counter(k_labels).most_common(1)[0][0])
        return np.array(predictions)

# Compare Euclidean vs Manhattan
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
metrics = ['euclidean', 'manhattan']

for ax, metric in zip(axs, metrics):
    knn = KNNWithMetric(k=3, metric=metric)
    knn.fit(X, y)
    
    Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    ax.contourf(xx, yy, Z, alpha=0.3, colors=['blue', 'red'])
    ax.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap='bwr', marker='o', s=30)
    ax.set_title(f'Metric: {metric.capitalize()} (k=3)')
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
