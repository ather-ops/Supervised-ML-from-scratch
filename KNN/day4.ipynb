## Step 1: Importing Libraries

Counter is used to count the occurrences of elements in a list or iterable. In KNN, after finding the k nearest neighbor labels, Counter helps count how many times each label appears.

```python
import numpy as np
from collections import Counter

print("="*50)
print("KNN DAY 4 - IMPLEMENTING KNN FROM SCRATCH")
print("="*50)
```

## Step 2: Defining the Euclidean Distance Function

```python
def euclidean_distance(point1, point2):
    """
    Calculate Euclidean distance between two points.
    
    Parameters:
    point1, point2: Lists or numpy arrays
    
    Returns:
    float: Euclidean distance
    """
    return np.sqrt(np.sum((np.array(point1) - np.array(point2)) ** 2))

# Test the function
p1 = [1, 2]
p2 = [4, 6]
print(f"Euclidean distance between {p1} and {p2}: {euclidean_distance(p1, p2):.4f}")
```

## Step 3: KNN Prediction Function

```python
def knn_predict(training_data, training_labels, test_point, k):
    """
    Predict the class of a test point using KNN algorithm.
    
    Parameters:
    training_data: List of training points
    training_labels: List of labels for training points
    test_point: Point to predict
    k: Number of nearest neighbors
    
    Returns:
    Predicted label
    """
    # Step 1: Calculate distances to all training points
    distances = []
    for i in range(len(training_data)):
        dist = euclidean_distance(test_point, training_data[i])
        distances.append((dist, training_labels[i]))
    
    # Step 2: Sort by distance (nearest first)
    distances.sort(key=lambda x: x[0])
    
    # Step 3: Get labels of k nearest neighbors
    k_nearest_labels = [label for _, label in distances[:k]]
    
    # Step 4: Majority voting
    return Counter(k_nearest_labels).most_common(1)[0][0]
```

```python
# Training data
training_data = [[1, 2], [2, 3], [3, 4], [6, 7], [7, 8]]
training_labels = ['A', 'A', 'A', 'B', 'B']

# Test point
test_point = [4, 5]
k = 3

print("Training Data:")
for i, (point, label) in enumerate(zip(training_data, training_labels)):
    print(f"  Point {i+1}: {point} -> Class {label}")
print(f"\nTest Point: {test_point}")
print(f"k = {k}")
```

```python
# Make prediction
prediction = knn_predict(training_data, training_labels, test_point, k)
print("\n" + "="*40)
print(f"Prediction for test point {test_point}: Class {prediction}")
print("="*40)
```

```python
def knn_predict_detailed(training_data, training_labels, test_point, k):
    """
    KNN prediction with detailed step-by-step output
    """
    print("\n" + "="*50)
    print("KNN PREDICTION - DETAILED STEPS")
    print("="*50)
    
    # Step 1: Calculate distances
    print("\nStep 1: Calculating distances to all training points")
    distances = []
    for i in range(len(training_data)):
        dist = euclidean_distance(test_point, training_data[i])
        distances.append((dist, training_labels[i]))
        print(f"  Distance to point {training_data[i]} (Class {training_labels[i]}): {dist:.4f}")
    
    # Step 2: Sort
    print("\nStep 2: Sorting by distance")
    distances.sort(key=lambda x: x[0])
    for i, (dist, label) in enumerate(distances):
        print(f"  Rank {i+1}: Distance = {dist:.4f}, Label = {label}")
    
    # Step 3: Get k nearest
    print(f"\nStep 3: Selecting k={k} nearest neighbors")
    k_nearest_labels = [label for _, label in distances[:k]]
    for i, label in enumerate(k_nearest_labels):
        print(f"  Neighbor {i+1}: Label = {label}")
    
    # Step 4: Voting
    print("\nStep 4: Majority voting")
    counter = Counter(k_nearest_labels)
    print(f"  Vote counts: {dict(counter)}")
    prediction = counter.most_common(1)[0][0]
    print(f"  Majority class: {prediction}")
    
    return prediction
```

```python
# Run detailed prediction
prediction_detailed = knn_predict_detailed(training_data, training_labels, test_point, k)
```

```python
print("\n" + "="*50)
print("Testing Different k Values")
print("="*50)

k_values = [1, 2, 3, 4, 5]

for k_val in k_values:
    pred = knn_predict(training_data, training_labels, test_point, k_val)
    print(f"k = {k_val}: Prediction = {pred}")
```

```python
class KNN:
    """
    Complete KNN implementation from scratch
    """
    
    def __init__(self, k=3):
        self.k = k
        self.X_train = None
        self.y_train = None
    
    def fit(self, X, y):
        """Store training data"""
        self.X_train = np.array(X)
        self.y_train = np.array(y)
    
    def _euclidean_distance(self, x1, x2):
        return np.sqrt(np.sum((x1 - x2) ** 2))
    
    def predict(self, X):
        """Predict for multiple test points"""
        X = np.array(X)
        predictions = []
        
        for x in X:
            distances = []
            for i in range(len(self.X_train)):
                dist = self._euclidean_distance(x, self.X_train[i])
                distances.append((dist, self.y_train[i]))
            
            distances.sort(key=lambda d: d[0])
            k_labels = [label for _, label in distances[:self.k]]
            predictions.append(Counter(k_labels).most_common(1)[0][0])
        
        return np.array(predictions)
    
    def score(self, X, y):
        """Calculate accuracy"""
        return np.mean(self.predict(X) == np.array(y))
```

```python
# Test the complete class
knn = KNN(k=3)
knn.fit(training_data, training_labels)

# Predict on training data
predictions = knn.predict(training_data)
accuracy = knn.score(training_data, training_labels)

print("Predictions on training data:")
for point, pred, actual in zip(training_data, predictions, training_labels):
    print(f"  {point} -> Predicted: {pred}, Actual: {actual}")
print(f"\nAccuracy: {accuracy*100:.2f}%")
```
