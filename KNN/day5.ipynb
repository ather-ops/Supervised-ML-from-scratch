# K-Nearest Neighbor (KNN) - Day 5 (From Scratch)

## Applications of KNN

### Recommendation Systems
Suggests items like movies or products by finding users with similar preferences.

### Spam Detection
Identifies spam emails by comparing new emails to known spam and non-spam examples.

### Customer Segmentation
Groups customers by comparing their shopping behavior to others.

### Speech Recognition
Matches spoken words to known patterns to convert them into text.

---

## Advantages of KNN

| Advantage | Description |
|-----------|-------------|
| Simple to use | Easy to understand and implement |
| No training step | Just stores data and uses it during prediction |
| Few parameters | Only needs k and distance method |
| Versatile | Works for both classification and regression |

---

## Disadvantages of KNN

| Disadvantage | Description |
|--------------|-------------|
| Slow with large data | Compares every point during prediction |
| Struggles with many features | Accuracy drops with too many features |
| Can Overfit | Overfits when k is very small |
| Curse of Dimensionality | Distance metrics become less meaningful in high dimensions |

---

## One-Line Summary

**KNN is simple and versatile but struggles with large data, many features, and the curse of dimensionality.**

## Demonstrating Curse of Dimensionality

```python
def knn_accuracy_vs_dimensions():
    """
    Show how accuracy drops with increasing dimensions
    """
    np.random.seed(42)
    n_samples = 100
    dimensions = [2, 5, 10, 20, 50, 100]
    accuracies = []
    
    for dim in dimensions:
        # Create synthetic data
        X = np.random.randn(n_samples, dim)
        y = (X[:, 0] + X[:, 1] > 0).astype(int)
        
        # Split
        split = int(0.8 * n_samples)
        X_train, X_test = X[:split], X[split:]
        y_train, y_test = y[:split], y[split:]
        
        # Simple KNN from scratch
        from collections import Counter
        predictions = []
        for x in X_test:
            distances = [np.linalg.norm(x - xt) for xt in X_train]
            k_indices = np.argsort(distances)[:3]
            k_labels = y_train[k_indices]
            predictions.append(Counter(k_labels).most_common(1)[0][0])
        
        acc = np.mean(np.array(predictions) == y_test)
        accuracies.append(acc)
        print(f"Dimensions = {dim:3}: Accuracy = {acc:.4f}")
    
    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(dimensions, accuracies, marker='o', linewidth=2)
    plt.xlabel('Number of Dimensions')
    plt.ylabel('Accuracy')
    plt.title('Curse of Dimensionality: Accuracy vs Dimensions')
    plt.grid(alpha=0.3)
    plt.show()

knn_accuracy_vs_dimensions()
```

```python
# Create noisy dataset
np.random.seed(42)
X_noisy = np.random.randn(200, 2)
y_noisy = (X_noisy[:, 0] + X_noisy[:, 1] > 0).astype(int)
X_noisy[:10] += np.random.randn(10, 2) * 0.5  # Add noise

# Compare k=1 vs k=5
def simple_knn(X_train, y_train, X_test, k):
    predictions = []
    for x in X_test:
        distances = [np.linalg.norm(x - xt) for xt in X_train]
        k_indices = np.argsort(distances)[:k]
        k_labels = y_train[k_indices]
        predictions.append(Counter(k_labels).most_common(1)[0][0])
    return np.array(predictions)

# Split
split = int(0.8 * len(X_noisy))
X_train, X_test = X_noisy[:split], X_noisy[split:]
y_train, y_test = y_noisy[:split], y_noisy[split:]

# Test different k
k_values = [1, 3, 5, 10, 20]
train_accs = []
test_accs = []

for k in k_values:
    pred_train = simple_knn(X_train, y_train, X_train, k)
    pred_test = simple_knn(X_train, y_train, X_test, k)
    train_accs.append(np.mean(pred_train == y_train))
    test_accs.append(np.mean(pred_test == y_test))

print("Overfitting Analysis:")
print("="*50)
for i, k in enumerate(k_values):
    print(f"k = {k:2}: Train Accuracy = {train_accs[i]:.4f}, Test Accuracy = {test_accs[i]:.4f}")

# Plot
plt.figure(figsize=(10, 6))
plt.plot(k_values, train_accs, marker='o', label='Train Accuracy', color='blue')
plt.plot(k_values, test_accs, marker='s', label='Test Accuracy', color='red')
plt.xlabel('k Value')
plt.ylabel('Accuracy')
plt.title('Overfitting: Small k Overfits, Large k Generalizes')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
import time

print("\n" + "="*50)
print("Prediction Time vs Dataset Size")
print("="*50)

sizes = [50, 100, 200, 500, 1000]
times = []

for size in sizes:
    # Create data
    X = np.random.randn(size, 2)
    y = (X[:, 0] + X[:, 1] > 0).astype(int)
    
    # Split
    split = int(0.8 * size)
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]
    
    # Time prediction
    start = time.time()
    for x in X_test:
        distances = [np.linalg.norm(x - xt) for xt in X_train]
        k_indices = np.argsort(distances)[:3]
        k_labels = y_train[k_indices]
        Counter(k_labels).most_common(1)[0][0]
    end = time.time()
    times.append(end - start)
    print(f"Size = {size:4}: Prediction time = {end - start:.4f} seconds")

# Plot
plt.figure(figsize=(10, 6))
plt.plot(sizes, times, marker='o', linewidth=2)
plt.xlabel('Dataset Size')
plt.ylabel('Prediction Time (seconds)')
plt.title('KNN: Prediction Time vs Dataset Size')
plt.grid(alpha=0.3)
plt.show()
```
